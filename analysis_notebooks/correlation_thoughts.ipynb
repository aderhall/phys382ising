```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib widget
```

```python
xdata = np.linspace(0, 1, 100)
# noise centered at zero
noise = np.random.normal(0, 0.1, 100)
m = 3.14
c = 2.71
ydata = m * xdata + c + noise
```

```python
(p_m, p_c), covm = curve_fit(lambda x, m, c: m * x + c, xdata, ydata)
corrm = covm / np.outer(np.sqrt(np.diag(covm)), np.sqrt(np.diag(covm)))
print(p_m, p_c, corrm)
```

```python

fig, ax = plt.subplots()
ax.plot(xdata, ydata)
ax.plot(xdata, p_m * xdata + p_c)
ax.set_xlabel('x')
ax.set_ylabel('y')
```
